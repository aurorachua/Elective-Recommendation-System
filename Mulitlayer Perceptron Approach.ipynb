```python
import pandas as pd
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_excel('Elective.xlsx')
dataset.head()
```

```python
dataset.shape
```

```python
x = dataset.drop(['Student', 'Gender', 'MA1', 'MA2', 'BIO.1', 'BAFS (ACC)', 'BAFS (BM)', 'CHEM.1', 'EL', 'CH.1', 'CL', 'EC', 'GEO', 'HI', 'ICT.1', 'PHY.1', 'PE DSE', 'VA.1'], axis=1)
y = dataset['CHEM.1']
```

```python
print(x.head(2))
print(y.head(2))
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size = 0.3)
print(x_train.shape)
print(y_test.shape)
```

```python
model = MLPRegressor(hidden_layer_sizes=(100, 100,), activation = 'identity', max_iter= 500)
```

```python
model.fit(x_train, y_train)
```

```python
predicted = model.predict(x_test)
predicted
```

```python
y_test
```

```python
model.score(x,y)
```
